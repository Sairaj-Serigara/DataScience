```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

employees
```

```python
departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
departments
```

```python
pd.merge(employees,departments,on="DeptID") #merge employye and department on deptid where ddeptid values are common
                                            #inneer join
```

```python
#left join
pd.merge(employees,departments,on="DeptID",how="left") #prints each and every rows of left (employee),and common rows from 
                                                        #right(departments)
```

```python
#right join
pd.merge(employees,departments,on="DeptID",how="right") #prints each and every rows of right (department),and common rows 
                                                        #from left(employee)       
```

```python
#outer join
pd.merge(employees,departments,on="DeptID",how="outer")#eachand every values of both df
```
