```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
#melt

# df.melt(id_vars,value_vars,var_name,value_name)

# id_vars--> columns which should be as it is
# value_vars-->columns which need to be changed 
# var_names-->columns name of changed columns
# val_names-->var_name renames the "variable" column.
```

```python
df_long=df.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
df_long
```

```python
#pivot

df_long.pivot(index="Name", columns="Subject", values="Score")

# index="Name": The unique values in the "Name" column will become the rows in the new DataFrame.
# columns="Subject": The unique values in the "Subject" column will become the columns in the new DataFrame.
# values="Score": The values from the "Score" column will populate the table
```
