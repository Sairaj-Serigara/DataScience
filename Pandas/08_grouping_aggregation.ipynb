```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "z", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
df
```

```python
df.groupby("Department")["Salary"].mean()    #group pf department then take salary column
```

```python
type(df.groupby("Department")["Salary"].mean() )
```

```python
df.groupby("Team")["Salary"].count()  #group of team in that number of salary 
```

```python
df.groupby("Team")["Salary"].sum()
```

```python
df.groupby("Team")["Salary"].min()
```

```python
df.groupby("Team")["Salary"].max()
```

```python
df.groupby(["Department","Team"])["Salary"].mean()
```

```python
df.groupby("Department")["Salary"].agg(["mean","min","max"])   #group of department in that mean,,min of sal
```

```python
type(df.groupby("Department")["Salary"].aggregate(["mean","min","max"]))
```

```python
df.groupby("Department")["Salary"].agg(
    avg_sal="mean",
    max_sal="max")
```

```python
df["Team_avg"]=df.groupby("Department")["Salary"].transform("mean") #create a new col avg_sal of department ,in that avg of 
df                                                                  #sal
```

```python
df.groupby("Department").filter(lambda x:x["Salary"].mean()>85)
```
