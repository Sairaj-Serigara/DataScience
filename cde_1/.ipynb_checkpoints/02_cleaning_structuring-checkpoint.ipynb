```python
import json

def clean_data(data):
    #remove user with missing names
    data["user"]=[user for user in data["users"] if user["name"].strip()]
    print("data cleaned success")

    """clean_users = []

    for user in data["users"]:
        if user["name"].strip():
            clean_users.append(user)
    
    data["users"] = clean_users
    """

    #remove duplicate friends
    for user in data["users"]:
        user["friends"]=list(set(user["friends"])) #created a set to remove the duplicate element
    print("duplicate friends removed")


    #remove inactive user
    data["users"]=[user for user in data["users"] if user["friends"] or user["liked_pages"]]
    print("users with no friends and no liked pages removed")
    return data
    

data=json.load(open("data2.json"))
data=clean_data(data)
cleaned_data=json.dump(data,open("cleaned_data2.json","w"))
    
```
