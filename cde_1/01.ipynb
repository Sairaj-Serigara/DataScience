```python
#function to load data
import json
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=load_data("data.json")
```

```python
#function to display
def display_users(data):
    for i in data["users"]:
        print(f"ID.{i['id']}:{i['name']} is friend with {i['friends']} and liked pages {i['liked_pages']}")
    print("\n")

    print("Pages information")
    for j in data["pages"]:
        print(f"{j['id']}--->{j['name']}") 
```

```python
display_users(data)
```
