```python
import json

def load_data(filename):
    with open("data.json","r") as f:
        data=json.load(f)
        return data

def people_you_may_know(user_id,data):
    user_friends={}
    for i in data["users"]:
        user_friends[i["id"]]=set(i["friends"])

    if user_id not in user_friends:
        return
    
    direct_friends=user_friends[user_id]
    sugestions={}
    for i in direct_friends:
        for j in user_friends[i]:
            if j!=user_id and j not in direct_friends:
                sugestions[j]=sugestions.get(j,0)+1
    sugestions = sorted(sugestions.items(), key=lambda x: x[1], reverse=True)
    return sugestions


data=load_data("data.json")
user_id=3
rec=people_you_may_know(user_id,data)
print(rec)
```
