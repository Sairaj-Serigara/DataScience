```python
with open("initialdata.txt",encoding="utf-8") as f:
    data=f.read()
data
```

```python
chunks=data.split("\n\n")
chunks
```

```python
def parse_chunk(chunk):
    chunk=chunk.strip()
    sep_chunk=chunk.split("\n")
    username=sep_chunk[0]
    posts=int(sep_chunk[1].split(" post")[0].replace(",",""))   # post: 340 s  ,to take only 340 we have to write [0[
    followers=float(sep_chunk[2].split(" followers")[0].replace("K","").replace("M",""))
    if "K" in sep_chunk[2]:
        followers=int(followers*1000)

    elif "M" in sep_chunk[2]:
        followers=int(followers*1000000)

    else:
        followers=followers
    following=int(sep_chunk[3].split(" following")[0].replace("K","").replace("M",""))
    if "K" in sep_chunk[3]:
        following=int(following*1000)

    elif "M" in sep_chunk[3]:
        following=int(following*1000000)

    else:
        following=following
    bio=sep_chunk[4:]
    # print(username,posts,followers,following,bio,sep="\n")
    return{"username":username,"posts":posts,"followers":followers,"following":following,"bio":bio}
    
# print(parse_chunk(chunks))
    
```

```python
all_chunks=[]
for chunk in chunks:
    parsed_chunks=parse_chunk(chunk)
    all_chunks.append(parsed_chunks)

print(all_chunks)
```

```python
import json
res=json.dumps(all_chunks, indent=4)
print(res)
```

```python
#max post
max=0
for chunk in all_chunks:
    if(max<chunk['posts']):
        max=chunk['posts']
        chunk_with_maxpost=chunk
print(chunk_with_maxpost)
```

```python
#max followers
max=0
for chunk in all_chunks:
    if(max<chunk['followers']):
        max=chunk['followers']
        chunk_with_followers=chunk
print(chunk_with_followers)
```

```python
#max followwing
max=0
for chunk in all_chunks:
    if(max<chunk['following']):
        max=chunk['following']
        chunk_with_following=chunk
print(chunk_with_following)
```
