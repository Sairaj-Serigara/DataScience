```python
import matplotlib.pyplot as plt
```

```python

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years,runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
import numpy as np

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))  # index positions
width = 0.25

plt.bar(sachin,sehwag,kohli)
plt.tight_layout()
```

```python
#$side by side plots

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x=np.arange(len(years))   #---->[0,1,2,3,4,5,6,7,8]

plt.bar(x-width,sachin,width=width,label="sachin")      #-->[0-0.8,1-0.8]
plt.bar(x,sehwag,width=width,label="sehwag")
plt.bar(x+width,kohli,width=width,label="kohli")

plt.xlabel("years")
plt.ylabel("runs")
plt.title("runs scored")
plt.legend()
plt.xticks(x,years)  #replace x with years
plt.tight_layout()   #is used to automatically adjust spacing in a plot so that labels, titles, and ticks don't overlap or get cut off.
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]

plt.barh(players,runs_5yrs, color="skyblue")   #horizontal bar
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
plt.show()
```

```python
# plt.text(x,y,text)
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players,runs,color="pink")

for i in range(len(players)):    #0,1,2   0-sachin,1=sehwag,2-kohli
    plt.text(i,runs[i]+50,str(runs[i]),ha="center")   #i=0 th pos,runs[i]=1500+50,1500 

plt.ylabel("runs")
plt.xlabel("player")
```

```python
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.barh(players,runs,color="pink")

for i in range(len(players)):    #0,1,2   0-sachin,1=sehwag,2-kohli
    plt.text(runs[i]+50,i,str(runs[i]),ha="center")   #i=0 th pos,runs[i]=1500+50,1500 
plt.tight_layout()
plt.ylabel("player")
plt.xlabel("run")
```
