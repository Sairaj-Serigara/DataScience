```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')

labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("runs scored")
plt.pie(runs,labels=labels)
```

```python
colors=['#181894','#99dfd5','#6fb637','#ee1b57']
plt.pie(runs,labels=labels,colors=colors)
```

```python
plt.pie(runs,
        labels=labels,
        colors=colors,
        wedgeprops={"edgecolor":"blue",'linewidth':2,"linestyle":"--",'fill':False}
       )
```

```python
explode=[0.1,0,0,0.1]
plt.pie(runs,labels=labels,colors=colors,explode=explode,shadow=True)
plt.title("Exploded sachins score")
```

```python
explode=[0.1,0,0,0.1]
plt.pie(runs,labels=labels,colors=colors,explode=explode,shadow=True,startangle=180)
plt.title("Exploded sachins score")
```

```python
explode=[0.1,0,0,0.1]
plt.pie(runs,labels=labels,colors=colors,explode=explode,shadow=True,startangle=180,autopct='%1.2f%%')
plt.title("Exploded sachins score")
```
