```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
tips.info()
```

```python
import matplotlib.pyplot as plt
sns.lineplot(x="total_bill",y="tip",data=tips)
plt.title("lineplot")
```

```python
sns.scatterplot(x="size",y="tip",data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Boxplot of Total Bill per Day')
plt.show()
```

```python
flights=sns.load_dataset("flights")
flights
```

```python
var=flights.pivot(index="month",columns="year",values="passengers")
var
```

```python
sns.heatmap(var,annot=True,fmt='d',cmap="YlGnBu")   #fmt===fromat =int     ,colorbar is shown by default
```

```python
import pandas as pd
 
df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})

sns.barplot(x="age",y="salary",hue="gender",data=df)    
'''hue="gender"

This is important 👇

hue splits data based on gender
So for each age, you get two bars:
Male (M)
Female (F)'''
```
