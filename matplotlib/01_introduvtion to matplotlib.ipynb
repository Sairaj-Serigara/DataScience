```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.xlabel("years")
plt.ylabel("runs")
plt.title("sachin thendolkurs graph")
plt.tight_layout()
plt.plot(years,runs)
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kohli,label="virat kohli")
plt.plot(years,sehwag,label="sehwag")

plt.xlabel("years")
plt.ylabel("runs")
plt.title("performance")
plt.legend()
plt.show()
```

```python

#format string
plt.plot(years,kohli,"ro--",label="kohli")    #color marker linrtype
plt.plot(years,sehwag,'g^:',label="sehwag")    #color:green, marker ^,type dotted
plt.legend()
```
