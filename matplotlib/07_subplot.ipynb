```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]

plt.subplot(2,1,2)   #plt.subplot(rows, columns, position)
plt.plot(x,y1)
plt.title("double of x")

plt.subplot(2,1,1)
plt.plot(x,y2)
plt.title("square of x")
```

```python
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]

plt.subplot(1,2,1)   #plt.subplot(rows, columns, position)
plt.plot(x,y1)
plt.title("double of x")
plt.xlabel("number")
plt.ylabel("double")

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("square of x")
plt.xlabel("number")
plt.ylabel("square")
plt.tight_layout()
```

```python
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]
y3 = [i ** 0.5 for i in x]
y4 = [10 - i for i in x]

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("double")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("square")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title(1**0.5)

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("10-i")
plt.tight_layout()
```

```python
fig,axs=plt.subplots(1,2,figsize=(10,5))
axs[0].plot(x,y1)
axs[0].set_title("double")

axs[1].plot(x,y2)
axs[1].set_title("square")

fig.suptitle("my plots",fontsize=14)
fig.subplots_adjust(top=0.85)
fig.savefig('my_photos.png')
```

```python
fig,axs=plt.subplots(2,2)

axs[0,0].plot(x,y1)
axs[0,1].plot(x,y2)
axs[1,0].plot(x,y3)
axs[1,1].plot(x,y4)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
ys = [y1, y2, y3, y4]
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']

for i in range(2):
    for j in range(2):
        idx = i * 2 + j
        axs[i, j].plot(x, ys[idx])
        axs[i, j].set_title(titles[idx])

plt.tight_layout()
plt.show()




"""
i	j	idx
0	0	0
0	1	1
1	0	2
1	1	3
"""
```
