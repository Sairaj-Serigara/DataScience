```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60, 65, 75, 85, 10]

colors=['red' if i < 50 else 'green' for i in exam_scores]
size=[i*2 for i in exam_scores]

plt.scatter(study_hours,exam_scores,c=colors,s=size)
plt.grid()
plt.xlabel("study hours")
plt.ylabel("scores")
plt.title("study vs marks")
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60, 65, 75, 85, 10]

plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='viridis')   #c=exam_scores----> color depends on exam_score
plt.colorbar(label="Score")   #must be written after the plot
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60, 65, 75, 85, 10]
scores=[84,12,56,4,1,99,13,75,14]
plt.scatter(study_hours,exam_scores,c=scores,cmap="binary")  #c is based on Scores
plt.colorbar(label='score')
plt.xlabel("study_hour")
plt.ylabel("score")
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60, 65, 75, 85, 10]
plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='viridis')   
plt.colorbar(label="Score")
for i in range(len(exam_scores)):
    plt.annotate(f's{i+1} ',(study_hours[i]+0.2,exam_scores[i]))
```
