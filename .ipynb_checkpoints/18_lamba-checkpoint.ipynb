```python
# lambda arguments : expression
```

```python
square=lambda x:x**2
```

```python
square(5)
```

```python
s=lambda a,b:a+b
s(8,2)
```

```python
numbers=[1,2,3,4,5,6]
squared=list(map(lambda i:i*i,numbers))  #map() gives a map object, not a list.
squared
```

```python
filterr=list(filter(lambda i:i%2 ==0,numbers))
filterr
```

```python
from functools import reduce
reducee=reduce(lambda x,y:x*y,numbers)
reducee
```

```python
add = lambda x, y: x + y
print(add(3, 7))  
```

```python
maximum = lambda x, y: x if x > y else y
print(maximum(10, 5))  
```

```python
students = [("Alice", 85), ("Bob", 78), ("Charlie", 92)]
students.sort(key=lambda student: student[1])  # Sort by score,,student[1] represents second value from tuple
print(students)
```
