```python
class smvitm:
    man="SMVITM"
    def __init__(self,princi,hod):
        self.princi=princi
        self.hod=hod

    def details(self):
        print(f"The priciple and hod of smvitm is {self.princi} and {self.hod}")

    @staticmethod    #no need to use self 
    def dean():
        print("dean is ")

    @classmethod
    def management(cls):
        print(f"management is {cls.man}")   """& {cls.princi} is not callable bcz it is not a cls var it is instance var && in classmethod if there
        is no class var it will not take instace variable   """

obj=smvitm("nagaraj","sada")
obj.details()
obj.dean()
obj.management()
```
